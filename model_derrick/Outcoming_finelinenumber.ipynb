```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
fl_dummies = pd.get_dummies(train_df['FinelineNumber'])
fl_dummies.tail()
```

```python
testing_df = train_df['FinelineNumber'].value_counts().reset_index()
testing_df.index = testing_df['index']
testing_df.drop(columns='index',inplace=True)
testing_df['ratio'] = testing_df['FinelineNumber'] * 100 / len(train_df['FinelineNumber'])
testing_df
```

```python
%%time
ls = []
for i in testing_df.index:
    ls.append(len(train_df[['DepartmentDescription','FinelineNumber']][train_df['FinelineNumber'] == i]))
len(ls)
```

```python
ls1 = []
for i in testing_df.index:
    ls1.append(i)
```

```python
df = pd.DataFrame()
df['FinelineNumber'] = ls1
df['Duplicated_num'] = ls
```

```python
df.tail()
```

```python
df['Duplicated_num'].describe()
```

```python
dfls1 = df[df['Duplicated_num']<10]['FinelineNumber'].values
len(dfls1)
```

```python
dfls2 = df[df['Duplicated_num']>115]['FinelineNumber'].values
len(dfls2)
```

```python
dfls = list(dfls1) + list(dfls2)
```

```python
dfls1 = dfls[:641]
dfls2 = dfls[641:1282]
dfls3 = dfls[1282:1923]
dfls4 = dfls[1923:]
```

```python
fl_dummies2 = fl_dummies[dfls2]
```

```python
fl_dummies3 = fl_dummies[dfls3]
```

```python
fl_dummies4 = fl_dummies[dfls4]
```
