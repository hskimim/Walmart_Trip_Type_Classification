```python
def unique_for_tt(df,comp='DepartmentDescription'):
    '''
    df : train.csv
    for : DD or FinelineNumber or Upc
    '''
    df = df[['TripType',str(comp)]]
    df_new = pd.DataFrame()
    df_new['TripType'] = df['TripType'].unique()
    df_new['unique_for_triptype'] = \
    [len(df[str(comp)][df['TripType']==i].unique()) for i in df['TripType'].unique()]
    return df_new.sort_values(by='TripType').reset_index(drop=True)
```

```python
testing_df = unique_for_tt(train_df)
```

```python
testing_df.sort_values(by='unique_for_triptype',ascending=False)
```
